## Определение яркости отраженного излучения

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Физические константы
H = 1.054_571_817e-34
C = 299_792_458
```

```python
def calculateQ(wave_len : int) -> float:
    return (H * C) / wave_len
```

```python
# Функция для расчета светлоты пискеля
def calculateFlux(wave_len : int, t_sec : int) -> float:
    return calculateQ(wave_len) / t_sec
```

```python
# Функция для расчета I
def calculateRadiantIntensity(wave_len : int, dw : float) -> float:
    return calculateFlux(wave_len, 1) / dw
```

```python
# Функция для расчета E (облученности)
def calculateIrradiance(wave_len : int, r : float) -> float:
    return calculateFlux(wave_len, 1) / (4 * math.pi * r * r)
```

```python
# Функция для расчета L (энергитическая яркость)
def calculateSurfaceRadiance(wave_Len : int, dA : float, dw : float, theta : float) -> float:
    F = calculateFlux(wave_Len, 1)
    return F**2 / (dA * math.cos(theta) * dw)
```

```python
L = calculateSurfaceRadiance(370, 10, 1, 0.5)
E = calculateIrradiance(370, 1)
BRDF = L / E

print(f"BRDF = {BRDF}")
```
