# BRDF

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
wave_lens = np.linspace(370, 730, (730 - 370))
```

```python
phi = np.exp(-((wave_lens - 600) ** 2) / (2 * 50 ** 2))
```

```python
r = 2
Ei = phi / (4 * np.pi * r ** 2)
c = 0.55 + 0.35 * np.sin(2 * np.pi * (wave_lens - 370) / (730 - 370))
```

```python
brdf = c / np.pi
```

```python
lo = brdf * Ei
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(3, 1, 1)
plt.plot(wave_lens, Ei)
plt.xlabel("Длина волны, нм")
plt.ylabel("Облученность, Вт/м2")
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(wave_lens, c)
plt.xlabel("Длина волны, нм")
plt.ylabel("Альбедо")
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(wave_lens, lo)
plt.xlabel("Длина волны, нм")
plt.ylabel("Яркость, Вт/(м2/ср)")
plt.grid()
```
