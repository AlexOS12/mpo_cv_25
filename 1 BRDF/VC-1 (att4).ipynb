```python
import numpy as np
from matplotlib import pyplot as plt
from math import radians, cos, pi
```

```python
# физические константы
H = 1.054_571_817e-34
C = 299_792_458
```

```python
# градусы -> радианы -> стерадианы
DW = radians(45) / 1.144 
DA = 5
R = 2 # растояние от источника света до плоскости по прямой под углом Theta к нормали плоскости
Theta = radians(30)
DW, R, Theta
```

```python
# Расчет Q
def calculate_Q(l : int) -> float:
    return H * C / l
```

```python
# Расчет потока Ф (Flux)
def calculate_flux(dq : float, dt : float) -> float:
    return dq / dt
```

```python
# Расчет силы излучения (Radiant Intensity)
def calculate_radiant_intensity(dF : float) -> float:
    return dF / DW
```

```python
# расчет облученности
def calculate_irradiance(flux : float) -> float:
    return flux /  (4 * pi * (R ** 2))
```

```python
# Расчет энергитической яркости (Surface Radiance)
def calculate_surface_radiance(flux : float) -> float:
    return flux ** 2 / (DA * cos(Theta) * DW)
```

```python
wave_lens = list(range(370, 730, 20))
Lo = []
Ei = []

for wave_len in wave_lens:
    l = calculate_surface_radiance(calculate_flux(calculate_Q(wave_len), 1.0))
    Lo.append(l)
    Ei.append(calculate_irradiance(calculate_flux(calculate_Q(wave_len), 1.0)))
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(wave_lens, Lo)
plt.subplot(2, 1, 2)
plt.plot(wave_lens, Ei)
```
