# BRDF

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
wave_lens = np.arange(370, 730, 5)
```

```python
phi = np.exp(-((wave_lens - 600) ** 2) / (2 * 50 ** 2))
```

```python
r = 2
Ei = phi / (4 * np.pi * r ** 2)
# альбедо условно соотвествует обычной белой бумаге
c = 0.75 + 0.15 * np.sin(2 * np.pi * (wave_lens - 370) / (730 - 370)) 
```

```python
brdf = c / np.pi
```

```python
lo = brdf * Ei
```

```python
# Функция для построения графика
def draw_plot(dataX, dataY,  xlabel, ylabel):
    plt.figure(figsize=(10, 5))
    plt.plot(dataX, dataY)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid()
```

```python
draw_plot(wave_lens, Ei, "Длина волны, нм", "Облученность, Вт/м2")
draw_plot(wave_lens, lo, "Длина волны, нм", "Яркость, Вт/(м2ср)")
draw_plot(wave_lens, c,  "Длина волны, нм", "Альбедо")
```

```python
# Входные данные
dAs = 1 # площадь источника света
D = 5   # диаметр линзы
F = 1   # растояние от линзы до приемника
Z = 1   # растояние от плоскости до линзы
Alpha = np.deg2rad(30)
Theta = np.deg2rad(30)
```

```python
dw_L = ((np.pi * D ** 2) / 4) * (np.cos(Alpha) / ((Z / np.cos(Alpha)) ** 2))
E_pr = lo * dw_L
```

```python
draw_plot(wave_lens, E_pr, "Длина волны, нм", "Облученность приемника, Вт/ср^2")
draw_plot(wave_lens, Ei, "Длина волны, нм", "Облученность, Вт/м2")
```
